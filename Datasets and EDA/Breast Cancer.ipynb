```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## 导入数据集
data = load_breast_cancer()

## 微改数据集
# 特征
X = pd.DataFrame(data.data, columns = data.feature_names)
# 乳腺癌情况
Y = list(data.target)
# 重组数据集
breast_cancer = X
breast_cancer['cancer'] = Y
breast_cancer
```

```python
# 描述性统计
malignant = breast_cancer[breast_cancer['cancer'] == 1]
benign = breast_cancer[breast_cancer['cancer'] == 0]
breast_cancer.describe()
#malignant.describe()
#benign.describe()
```

```python
# 相关系数矩阵
breast_cancer.corr()
#malignant.corr()
#benign.corr()
```

```python
# 热力图
sns.heatmap(breast_cancer.corr())
```
